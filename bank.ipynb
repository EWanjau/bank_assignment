##  Bank application
### The application allows a user to:
``` Deposit money
    Withdraw money
    Check balance
    Exit program
```

````python
from account import Account

# Create account
user_name = input("What is your Name: ")
user1 = Account(user_name)

while True:
    user_choice = input(f"""
Welcome {user1.name}, to Lux Bank:
What would you like to do(1-4)?
1. Check Balance
2. Deposit Cash
3. Withdraw Cash
4. Exit Service""")
    choice = user_choice.strip()
    if choice == '1':
        print("Your Account Balance is: Kshs", user1.check_balance())

    elif choice == '2':
        dep_amount = int(input("Enter Amount to Deposit: Kshs "))
        if dep_amount <= 0:
                print("Deposit must be Kshs 1 and above")
        else:
            new_balance = user1.deposit_amount(dep_amount)
            print(f"You have deposited Kshs {dep_amount}")
            print(f"New Account Balance: Kshs {new_balance}")

    elif choice == '3':
        withdraw_cash = int(input("Enter the amount you wish to withdraw: Kshs "))
        if withdraw_cash > user1.amount:
            print("Sorry, you have insufficient balance.")
            print(f"Your Account Balance is: Kshs {user1.amount}")
        elif withdraw_cash <= 0:
            print("Amount must be more than Kshs 1")
        else:
            new_balance = user1.withdraw_amount(withdraw_cash)
            print(f"You have withdrawn Kshs {withdraw_cash}")
            print(f"New Account Balance: Kshs {new_balance}")

    elif choice == '4':
        print("Thank you for banking with Lux Bank!")
        break

    else:
        print("You have entered an invalid option. Try again (1-4).")
````
