##  Bank application
### The application allows a user to:
``` Deposit money
    Withdraw money
    Check balance
    Exit program
```

````python
class Account():

    def __init__(self, name, amount=10000):
        self.name = name
        self.amount = amount

    def check_balance(self):
        return self.amount

    def deposit_amount(self, deposit):
        self.amount += deposit
        return self.amount

    def withdraw_amount(self, withdraw):
        self.amount -= withdraw
        return self.amount

user_name = input("What is your Name")
user1 = Account(user_name)
while True:
    user_choice = input(f"""Welcome {user_name}, to Lux Bank:

                    What would you like to do?
                        1. Check Balance
                        2. Deposit Cash
                        3. Withdraw Cash
                        4. Exit
                        (Select 1, 2, 3, 4)
                   """)
    if user_choice == '1':
        print("Your Account Balance is: Kshs",user1.check_balance())
    elif user_choice == '2':
        dep_amount = int(input("Enter Amount to Deposit: Kshs "))
        print(f"You have deposited Kshs {dep_amount}")
        new_balance = user1.deposit_amount(dep_amount)
        print(f"New Account Balance: Kshs {new_balance}")
    elif user_choice == '3':
        withdraw_cash = int(input("Enter the amount you wish to withdraw: KShs "))
        if withdraw_cash > user1.amount:
            print("Sorry you have insufficient balance")
            print(f"Your Account Balance is: {user1.amount}")
        else:
            print(f"You have withdrawn Kshs {withdraw_cash}")
            new_bal = user1.withdraw_amount(withdraw_cash)
            print(f"New Account Balance: {new_bal}")
    else:
        print("Thank you for banking with Lux Bank!")
        break
````
