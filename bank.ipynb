##  Bank application
### The application allows a user to:
``` Deposit money
    Withdraw money
    Check balance
    Exit program
```

````python
class Account():
    amount = 10000
    def __init__(self, name, deposit=0, withdraw=0):
        self.name = name
        self.deposit = deposit
        self.withdraw = withdraw

    def check_balance(self):
        return Account.amount

    def deposit_amount(self, deposit):
        self.amount += deposit
        return self.amount

    def withdraw_amount(self, withdraw):
        self.amount -= withdraw
        return self.amount

    def exit():
        pass

user_name = input("What is your Name")
user_choice = input(f"""Welcome {user_name} to Lux Bank: What would you like to do?
                        1. Check Balance
                        2. Deposit Cash
                        3. Withdraw Cash
                        (Select 1, 2, 3)
                   """)
if user_choice == '1':
    user1 = Account(user_name)
    print("Your Account Balance is: Kshs",user1.check_balance())
elif user_choice == '2':
    dep_amount = int(input("Enter Amount to Deposit: "))
    user1 = Account(user_name, deposit=deposit_amount)
    print(f"You have deposited {dep_amount}")
    new_balance = user1.deposit_amount(dep_amount)
    print(f"New Account Balance: {new_balance}")
else:
    withdraw_cash = int(input("Enter the amount you wish to withdraw: "))
    if withdraw_cash > Account.amount:
        print("Sorry you have insufficient balance")
        print(f"Your Account Balance is: {Account.amount}")
    else:
        user1 = Account(user_name, withdraw=withdraw_cash)
        print(f"You have withdrawn Kshs {withdraw_cash}")
        new_bal = user1.withdraw_amount(withdraw_cash)
        print(f"New Account Balance: {new_bal}")
````
